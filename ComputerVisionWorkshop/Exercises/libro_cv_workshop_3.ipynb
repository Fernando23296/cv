### Working with Histograms 

#### Histogramas
Un histograma de una imagen es una representación gráfica de la distribucion de las intensidades del pixel. La intensidad del pixel de una imagen tiene un rango de 0 a 255.

```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("../img/gato.jpg")
```

```python
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
# LAB es mas preciso en terminos de imitar al ojo humano. Usa L,A,B para especificar un color
img_lab = cv2.cvtColor(img, cv2.COLOR_LAB2RGB)
plt.imshow(img_lab)
plt.show()
```

En LAB:
<ul>
<li> El canal L (plano 0) tiene informacion sobre la iluminación </li>
    <li> EL canal A (plano 1) tiene rango entre verde y rojo </li>
    <li> El canal B (plano 2) tiene rango entre azul y amarillo</li>
</ul>

Entonces, entre los canales A y B se define la cromocidad(?).

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
plt.show()
```

#### Histogramas con Matplotlib

```python
img = cv2.imread("../img/river_scene.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow('',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ax = plt.hist(img.ravel(), bins = 256)
```

```python
img_2 = cv2.imread("../img/half.png", cv2.IMREAD_GRAYSCALE)
cv2.imshow('', img_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ax_2 = plt.hist(img_2.ravel(), bins = 256)
```

### Usando 10 bins

```python
ax_3 = plt.hist(img.ravel(), bins = 10)
```

#### Usando 2 bins

### Histograma en un ejemplo de color

```python
img = cv2.imread("../img/sunflower.jpg")
```

```python
im = img[:,:,2]
plt.hist(im.ravel(), bins = 256, color = "Red", alpha = 0.5)
im = img[:,:,1]
plt.hist(im.ravel(), bins = 256, color = "Green", alpha = 0.5)
im = img[:,:,0]
plt.hist(im.ravel(), bins = 256, color = "Blue", alpha = 0.5)
plt.show()
```
