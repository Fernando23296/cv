### Ejercicio, Water Effect

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../img/data/lion.jpg")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
img.shape
```

```python
imgNew = np.zeros((814, 640, 3), dtype = np.uint8)
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
imgNew[:407][:] = img
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
imgInverted = img[::-1,:,:]
plt.imshow(imgInverted[:,:,::-1])
plt.show()
```

```python
imgNew[407:][:] = imgInverted
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
cv2.imwrite("WaterEffect.png", imgNew)
```

### Ejercicio, Mirror Negative

```python
img2 = cv2.imread("../img/data/lion.jpg")
```

```python
plt.imshow(img2[:,:,::-1])
plt.show()
```

```python
img2.shape
```

```python
hsv_image = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image[:,:,::-1])
```

```python
h, s, v =  cv2.split(hsv_image)
plt.imshow(v, cmap = "gray")
plt.show()
```

```python
v = 255 - v
plt.imshow(v, cmap = "gray")
plt.show()
```

```python
img_hsv = cv2.merge((v,v,v))
plt.imshow(img_hsv[:,:,::-1])
plt.show()
```

```python
img_hsv = img_hsv[:,::-1,:]
plt.imshow(img_hsv[:,:,::-1])
plt.show()
```

```python
imgNew = np.zeros((img_hsv.shape[0], img_hsv.shape[1]*2, 3), dtype = np.uint8)
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
imgNew[:,:640,:] = img
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
imgNew[:,640:,:] = img_hsv
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
cv2.imwrite("../img/mirror.jpg", imgNew)
```
