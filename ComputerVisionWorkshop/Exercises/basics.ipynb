```python
# Para resize
resize_img = img.resize((200, 200))
resize_img.show()
```

```python
#para mejorar la imagen
from PIL import ImageEnhance
```

```python
# le da mas brillo
enhancer = ImageEnhance.Brightness(img)
enhancer.enhance(0.5).show() #si el valor es mayor hay mas brillo, si es menor es menos brillo, si es 1  es lo mismo
```

```python
enhancer = ImageEnhance.Contrast(img)
enhancer.enhance(2).show() #la misma idea que la del brillo
```

```python
from skimage import data
```

```python
io.imshow(data.camera())
io.show()
```

```python
io.imshow(data.text())
io.show()
```

```python
# Para cambiar el color
from skimage import io, color
img = io.imread("../img/gato.jpg")
gray = color.rgb2gray(img)
io.imshow(gray)
io.show()
```

```python
hsv = color.rgb2hsv(img)
io.imshow(hsv)
io.show()
```

```python
img = data.astronaut()
img_hsv = color.rgb2hsv(img)
io.imshow(img_hsv)
io.show()
```

```python
#Modulo para dibujar figuras
import numpy as np
from skimage import io, draw
```

```python
img = np.zeros((100,100), dtype=np.uint8)
x, y = draw.circle(50, 50, 10)
img[x, y] = 1
io.imshow(img)
io.show()
```

```python
img = np.zeros((100, 100), dtype = np.uint8)
x, y = draw.ellipse(50, 50, 10, 20)
img[x, y] = 1
io.imshow(img)
io.show()
```

```python
img = np.zeros((100,  100), dtype = np.uint8)
r = np.array([10, 25, 80, 100])
c = np.array([10, 60, 40, 100])
x, y = draw.polygon(r, c)
img[x, y] = 1
io.imshow(img)
io.show()
```
